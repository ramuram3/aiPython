분 -> 시간

```python
total=int(input("몇 분 인가요? "))
hour = total//60
min = total%60
print("{}시간 {}분입니다".format(hour,min))
```

섭씨 -> 화씨

```python
cTemp=float(input("섭씨 몇 도 인가요? "))
fTemp = cTemp * 1.8 + 32
print("섭씨 %0.2f도, 화씨 %0.2f도 입니다"%(cTemp,fTemp))
```

원 둘레 면적

```python
import math
radius = float(input("반지름이 얼마인가요? "))
round = radius * math.pi * 2
area = radius**2 * math.pi
print("둘레 %0.2f, 면적 %0.2f입니다"%(round,area))
print("둘레 {:0.2f}, 면적 {:0.2f}입니다".format(round,area))
```

성적 판정

```python
while True:
    scoreIn = input("점수? :")
    if scoreIn.isdigit():
        score = int(scoreIn)
        if score >= 0 and score<= 100:
            break
        else:
            print("범위를 벗어난 숫자입니다")
            continue
    else:
        print("숫자만 입력")

print(score)
if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

홀짝 합

```python
oddSum = 0
evenSum = 0

for i in range(1,11):
    if i % 2 == 0:
        evenSum += i
    else:
        oddSum += i

print("홀수 합 = {}, 짝수 합 = {}".format(oddSum,evenSum))
```
